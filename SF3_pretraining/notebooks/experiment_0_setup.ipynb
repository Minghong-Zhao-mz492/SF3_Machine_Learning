```python
# Step 1

from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt

PROJECT_ROOT = Path("..").resolve()

DATA_DIR = PROJECT_ROOT / 'data'
FIGURE_DIR = PROJECT_ROOT / 'figures'
NOTEBOOK_DIR = PROJECT_ROOT / 'notebooks'
SRC_DIR = PROJECT_ROOT / 'src'

for folder in [DATA_DIR, FIGURE_DIR, NOTEBOOK_DIR, SRC_DIR]:
    folder.mkdir(exist_ok=True)

np.random.seed(0)

print('Project folders are ready.')
print('Data folder:', DATA_DIR)
print('Figure folder:', FIGURE_DIR)
print("Random seed fixed.")
```

```python
# Step 2

def save_figure(fig,filename,folder=FIGURE_DIR,dpi=200):
    folder.mkdir(exist_ok=True)
    path = folder / filename
    fig.savefig(path,dpi=dpi,bbox_inches='tight')

    print("Saved figure to:",path)
```

```python
x = np.linspace(-2,2,100)
y = x**2

x.shape,y.shape
```

```python
fig,ax = plt.subplots()
ax.plot(x,y)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_title('Test plot: $y=x^2$')

save_figure(fig, "exp0_test_plot.png")

fig,ax
```

```python
from datetime import date

LOG_PATH = PROJECT_ROOT / 'experiment_log.md'

def append_log_entry(log_path, title, purpose, parameters, results, next_step):
    today = date.today().isoformat()

    entry = f'''## {title}

Date: {today}

### Purpose
{purpose}

### Parameters
{parameters}

### Results
{results}

### Next step
{next_step}

---

'''
    print("Entry type:", type(entry))

    with open(log_path, 'a', encoding='utf-8') as f:
        f.write(entry)

    print("Appended log entry to:", log_path)
```

```python
append_log_entry(
    log_path=LOG_PATH,
    title = 'Experiment 0: Setup and coding discipline',
    purpose = '- Set up project folders\n- Fix Numpy random seed\n- Save figures\n- Start ab log',
    parameters="- Random seed: 0\n- Figure folder: figures/\n- Test plot: exp0_test_plot.png",
    results="- Created data/, figures/, notebooks/, src/\n- Saved figures/exp0_test_plot.png",
    next_step="- Check reproducibility of random numbers"
)
```
